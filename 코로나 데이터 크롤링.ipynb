```python
#import pandas as pd
#import numpy as np
from selenium import webdriver
from tqdm import tqdm_notebook # 진행률 표시
import requests
#import pickle
#import re
#import ast

from bs4 import BeautifulSoup 
from urllib.request import urlopen
import urllib
import time
```

```python
driver = webdriver.Chrome('chromedriver')
naver_news_title = []
naver_news_content = []

for n in tqdm_notebook(range(len(naver_news_link))):
    #print(n)
    
    try:
        driver.get(naver_news_link[n])
        #print(naver_news_link[n][idx])

    except:
        print("Timeout!")
        continue

    try:
        response = driver.page_source

    except :
        driver.switch_to_alert().accept()
        #print("게시글이 삭제된 경우입니다.")
        continue

    soup = BeautifulSoup(response, "html.parser")

    title = None
    doc = None

    try:
        title = soup.select_one(".article_info > .tts_head").get_text()
        doc = soup.select_one("._article_body_contents").get_text()
        doc = doc.replace("본문 내용","")
        doc = doc.replace("// TV플레이어","")
        doc = doc.replace("TV플레이어","")
        doc = doc.replace("// flash 오류를 우회하기 위한 함수 추가","")
        doc = doc.replace("function _flash_removeCallback() {}","")
        doc = doc.strip();

    except:
        title = ""
        doc = ""

    #print(title)
    if title : naver_news_title.append(title)
    if doc : naver_news_content.append(doc.replace('\n', ' '))

    time.sleep(2)
    #break;
    
    
print(naver_news_title[0])
print("==================================")
print(naver_news_content[0])
```
