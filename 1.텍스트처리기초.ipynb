### NLTK (Natural Language Tool Kit)
* 자연어 처리, 분석용 패키지

```python
from nltk.book import *
```

### corpus
* 말뭉치
* 단어, 문서 집합

```python
nltk.corpus.gutenberg.fileids()
```

```python
# 제인오스틴의 이성과 감성 (austen-sense.txt)
data = nltk.corpus.gutenberg.raw('austen-sense.txt')
print(data[:1500])
```

### token
* 문서내 문장, 단어로 구분하는 단위
* 토큰화(tokenizing) - 문서를 토큰으로 분리하는 작업

```python
# 문장
from nltk.tokenize import sent_tokenize
print(sent_tokenize(data[:1500])[1])
```

```python
# 단어
from nltk.tokenize import word_tokenize
word_tokenize(data[:1500])
```

```python
# 정규표현식
from nltk.tokenize import RegexpTokenizer
reg = RegexpTokenizer("[\w]+")
reg.tokenize(data[:1500])
```

### morph
* 형태소
* 의미가 있는 가장 작은 단위 문자
* 보통 토큰단위 형태소

### pos (Part Of Speech)
* 품사
* 문법 형태, 의미/뜻
* VB:동사, NN:명사

```python
from nltk.tag import pos_tag
sent = 'Their estate was large, and their residence was at Norland Park'
pos_tag(word_tokenize(sent))
```

```python
# 명사만 추출(토큰 NN)
## 일반 for문
nn_list = []
for t in pos_tag(word_tokenize(sent)):
    #print(t[0], t[1])
    if t[1] == 'NN' : nn_list.append(t[0])
nn_list
```

```python
## 내포된(comprehension) for문
nn_list = [t[0] for t in pos_tag(word_tokenize(sent)) if t[1] == 'NN']
nn_list
```

```python
# Text 클래스
from nltk import Text
text = Text(reg.tokenize(data))
```

```python
# 토큰 빈도 그래프
text.plot(20)
```

```python
# FreqDist 클래스
# 토큰별로 사용빈도를 저장하고 있는 클래스
# Text클래스의 vocab() 메서드로 호출
fd = text.vocab()
```

```python
type(fd)
```

```python
from nltk import FreqDist
token_list = pos_tag(reg.tokenize(data))
nnp_list = [t[0] for t in token_list if t[1] == 'NNP']
fd = FreqDist(nnp_list)
```

```python
# 전체 단어 수, 단어 빈도수, 확률
fd.N(), fd['Norland'], fd.freq('Norland')
```

```python
# 가장 빈도수가 많은 단어 5개
fd.most_common(5)
```
