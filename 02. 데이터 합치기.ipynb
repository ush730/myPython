```python
import pandas as pd
```

```python
df1 = pd.read_csv('../data/concat_1.csv')
df2 = pd.read_csv('../data/concat_2.csv')
df3 = pd.read_csv('../data/concat_3.csv')
```

```python
df3
```

```python
# concat 메서드로 데이터프레임 합치기
# 수직(위/아래)로 합쳐짐
concat_df = pd.concat([df1,df2,df3])
concat_df
```

```python
concat_df.loc[0] # 인덱스명 중복
```

```python
concat_df2 = concat_df.reset_index(drop=True)
concat_df2
```

```python
# 시리즈를 데이터프레임에 합치기
s = pd.Series(['n1','n2','n3','n4'])
# df1과 s를 합치기
pd.concat([df1, s]) # 기본값이 열로 추가(열이름이 없으므로)
```

```python
# 하나의 행으로 된 데이터프레임
df_onerow = pd.DataFrame([['n1','n2','n3','n4']], columns=['A','B','C','D'])
df_onerow
```

```python
pd.concat([df1, df_onerow])
```

```python
# 하나만 합치는 경우
df1.append(df_onerow)
```

```python
# concat메서드에서 인덱스 새로 지정
pd.concat([df1,df2,df3], ignore_index=True)
```

```python
# 수평으로 합치기(열방향)
# axis : 축 (0:행,1:열)
df_col = pd.concat([df1, df2, df3], axis=1)
df_col
```

```python
df_col['A']
```

```python
df_col = pd.concat([df1, df2, df3], axis=1, ignore_index=True)
df_col
```

```python
df_col.columns = ['A','B','C','D','E','F','G','H','I','J','K','L']
df_col
```

```python
# 열의 일부가 다른 경우
df1.columns = ['A','B','C','D']
df2.columns = ['E','F','G','H']
df3.columns = ['A','C','F','H']
df3
```

```python
# 그냥 concat하면 행기준, 없는 열은 NaN
pd.concat([df1, df2, df3])
```

```python
# df1, df3의 공통열(교집합)
pd.concat([df1, df3], join='inner')
```

```python
# 행기준으로 연결
df2.index=[4,5,6,7]
df3.index=[0,2,5,7]
```

```python
pd.concat([df1,df2,df3], axis=1)
```

```python
pd.concat([df1,df3], axis=1, join='inner')
```

* concat : 행이나 열을 기준으로 단순하게 이어 붙이기
* merge : 특정 키값을 기준으로 병합

### pd.merge(left, right, on='기준', how='병합방식')
* left, right 데이터프레임
* on 병합 기준이 되는 컬럼
* how : left, right, inner, outer 4가지 병합방식 지정

```python
df2_right = df2.drop([1,3,5,7,9]).reset_index(drop=True)
df2_right
```

```python
# 인덱스조건으로만 합쳐지기때문에 이름도 안맞고, NaN값도 생김
pd.concat([df,df2_right], axis=1)
```

```python
# df를 기준(왼쪽데이터프레임)으로 이름을 키값으로 병합
pd.merge(df, df2_right, on='이름', how='left')
```

```python
pd.merge(df, df2_right, on='이름', how='right')
```

```python
pd.merge(df, df2_right, on='이름', how='inner')
```

```python
pd.merge(df, df2_right, on='이름', how='outer')
```

```python
# 컬럼명이 다른 경우
df2_right.columns = ['성명', '연봉', '가족수']
df2_right
```

```python
pd.merge(df, df2_right, on='이름', how='left')
```

```python
# df에는 이름으로, df2_right에는 성명
pd.merge(df, df2_right, left_on='이름', right_on='성명', how='left')
```
