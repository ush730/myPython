```python
import tensorflow as tf
from tensorflow.keras.datasets import boston_housing
(train_x, train_y), (test_x, test_y) = boston_housing.load_data()
```

```python
train_x.shape, train_y.shape, test_x.shape, test_y.shape
```

```python
from sklearn.model_selection import train_test_split
train_x, valid_x, train_y, valid_y = train_test_split(train_x, train_y, random_state=10)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

tf.random.set_seed(14)

model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(13,)))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

model.fit(train_x, train_y, epochs=200, validation_data=(valid_x, valid_y))
```

```python
model.evaluate(test_x, test_y)
```

### 교차검증

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=5)
```

```python
# train_test_split은 하지않음
(train_x, train_y), (test_x, test_y) = boston_housing.load_data()
```

```python
# 모델구현 함수로 정의
def getModel():
    model = Sequential()
    model.add(Dense(64, activation='relu', input_shape=(13,)))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(16, activation='relu'))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

```python
tuple(kfold.split(train_x))
```

```python
'''
평가결과 저장할변수 선언
for (kfold갯수만큼)
학습데이터, 검증데이터 (kfold) 추출
모델 구현
학습(추출한데이터로)
평가(evaluate)
평가결과를 변수에 저장
변수의 평균
'''
maes = []
for train_idx, valid_idx in kfold.split(train_x):
    train_fold_x, valid_fold_x = train_x[train_idx], train_x[valid_idx]
    train_fold_y, valid_fold_y = train_y[train_idx], train_y[valid_idx]
    
    model = getModel()
    
    model.fit(train_fold_x, train_fold_y, epochs=200, validation_data=(valid_fold_x, valid_fold_y))
    
    loss, mae = model.evaluate(test_x, test_y)
    maes.append(mae)
```
