```python
import pandas as pd
```

```python
df = pd.read_csv('../data/온라인_판매기록.csv', encoding='ms949')
df
```

```python
df.info()
```

```python
# 평균을 구하는 함수, 소수점 두자리로 반올림
import numpy as np
def user_mean(col):
    avg = np.mean(col)
    return round(avg,2)
```

```python
# - 쇼핑몰 유형별로 제품들의 평균판매금액
```

```python
# - 쇼핑몰 유형별로 제품들의 평균판매금액
pd.pivot_table(df, values='판매금액', index='제품', columns='쇼핑몰 유형', aggfunc=lambda c : round(np.mean(c),2))
```

```python
# - 쇼핑몰 유형별로 제품들의 최대판매금액
pd.pivot_table(df, values='판매금액', index='제품', columns='쇼핑몰 유형', aggfunc='max')
```

```python
# - 쇼핑몰 유형별 판매수량 평균
df.groupby('쇼핑몰 유형')['수량'].mean()
```

```python
#- 쇼핑몰 유형별 판매금액의 평균과 최대값
df.groupby('쇼핑몰 유형')['판매금액'].agg(['mean','max'])
```

```python
#- 쇼핑몰 유형별 수량, 판매금액의 평균과 최대값
df.groupby('쇼핑몰 유형')[['수량','판매금액']].agg([lambda c : round(np.mean(c),2),'max'])
```
