### 주택도시보증공사_전국 신규 민간 아파트 분양가격 동향
### 전처리, 분석
* 데이터 다운로드
* 컬럼명 분양가격(㎡) -> 분양가격
* 결측치 처리
* 분양가격 int로 변경
* 규모구분 열에 '전용면적'이라는 문자열 제거
* 지역별 평균 분양가 출력
* 지역별 가장 비싼 분양가
* 연도별 평균 분양가

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20200430.csv', encoding='ms949')
df
```

### 숫자로된 값이 없는 경우
* 결측치
* '' (빈값), ' '(공백)
* 0
* -

### 값이 있는데, 변환이 안되는 경우
* ' 1000 '
* 1,000
* 통화기호가 들어간 경우 \1000, $1000

```python
# fillna : 결측치 채움
# dropna : 결측치 제거
df = df.dropna()
```

```python
df.info()
```

```python
# 컬럼명 분양가격(㎡) -> 분양가격
df = df.rename(columns={'분양가격(㎡)':'분양가격'})
```

```python
df.head()
```

```python
# 분양가격 int로 변경
df['분양가격'].astype(int)
```

```python
# 확인
df.loc[df['분양가격']=='  ']
```

```python
# 공백제거
df['분양가격']=df['분양가격'].str.strip()
```

```python
df['분양가격'].astype(int)
```

```python
# 확인
df.loc[df['분양가격']=='']
```

```python
df = df.dropna()
```

```python
# 0으로 바꾸자
# df.loc[df['분양가격']=='']['분양가격']=0
# (df.drop(df.loc[df['분양가격']==''].index)['분양가격'] == '').sum()
df = df.drop(df.loc[df['분양가격']==''].index)
```

```python
df.loc[df['분양가격']=='']
```

```python
df['분양가격'] = df['분양가격'].astype(int)
```

```python
df.info()
```

```python
# 7,311 -> ','를 제거하고 숫자로 변환
# - -> '-'를 '' 치환하고 '' 삭제, '-'를 삭제
```

```python
# 규모구분 열에 '전용면적'이라는 문자열 제거
df['규모구분'] = df['규모구분'].str.replace('전용면적 ','')
```

```python
df.head()
```

```python
# 지역별 평균 분양가 출력
df.groupby('지역명')['분양가격'].mean()
```

```python
# 지역별 가장 비싼 분양가
df.groupby('지역명')['분양가격'].max()
```

```python
# 연도별 평균 분양가
df.groupby('연도')['분양가격'].mean()
```

```python
# 연도별, 규모구분별 분양가격 평균
df.groupby(['연도','규모구분'])['분양가격'].mean()
```

```python
# 출력결과를 데이터프레임으로 출력
pd.DataFrame(df.groupby(['연도','규모구분'])['분양가격'].mean())
```
